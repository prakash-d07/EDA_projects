## lets divide the categorical and numerical data into two and store them into two lists

```python
df_categorical=[]
df_numerical=[]
df_datetime=[]
for i in  range(len(df.columns)):
    if df.iloc[:,i].dtype==np.dtype('O'):
        df_categorical.append(df.columns[i])
    elif df.iloc[:,i].dtype==np.dtype('int64'):
        df_numerical.append(df.columns[i])
    else:
        df_datetime.append(df.columns[i])
```
